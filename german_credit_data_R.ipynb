```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

# 1. descriptive statistics

```python
data = pd.read_csv("~/06_data/german_credit_data.xls")
data.head()
```

```python
plt.scatter(data["Duration"],data["Credit amount"], c = "blue", alpha = 0.5)
plt.xlabel("Duration")
plt.ylabel("Credit amount")
```

```python
idx1 = data[data["Credit amount"] > np.quantile(data["Credit amount"], 0.95)].index
idx2 = data.index.difference(idx1)

_ = sns.scatterplot(x = data.loc[idx1, "Duration"], y = data.loc[idx1, "Credit amount"],
                color = "r",alpha = 0.5)
_ = sns.scatterplot(x = data.loc[idx2, "Duration"], y = data.loc[idx2, "Credit amount"], 
                color = "b", alpha = 0.5)
        
```

```python
_ = sns.displot(data["Credit amount"])
```

```python
g = sns.boxplot(x = "Purpose", y = "Credit amount", hue = "Sex", data = data )
plt.xticks(rotation=30)
```
